# Lambda

```python
def str2int (x:str)->int:
  try:
    x = x.replace(',','')
    return(int(x))
  except:
    return(0)
```

```python
df_filmes['Gross'] = df_filmes['Gross'].apply(lambda x: str2int(x))

teste = df_filmes.groupby('Released_Year', as_index=False).agg({
      'Gross':['mean','sum'],
})

```

```python
teste.plot(x='Released_Year', y=[('Gross','mean')])
```

```python
teste.plot(x='Released_Year', y=[('Gross','sum')])
```
